```python
import pandas as pd

# Charger les données
name = 'Segbana'  # Nom du fichier sans l'extension
file_path = f'{name}.csv'  
data = pd.read_csv(file_path)


# Vérifier les premières lignes du fichie
data.head()
```

```python


# Calculer la somme glissante sur 6 mois
data['Moving_Sum_6'] = data['PRECTOTCORR_SUM'].rolling(window=6).sum()

# Calculer la moyenne et l'écart-type de la somme glissante
mean_6 = data['Moving_Sum_6'].mean()
std_6 = data['Moving_Sum_6'].std()

# Calculer le SPI6
data['SPI6'] = (data['Moving_Sum_6'] - mean_6) / std_6

# Afficher les premières lignes pour vérification
data.head()

```

```python
# Sauvegarder le fichier avec les nouvelles données SPI6
output_file_path = f'output/{name}_with_SPI6.csv'

data.to_csv(output_file_path, index=False)

print(f"Le fichier avec les données SPI6 a été sauvegardé sous : {output_file_path}")
```
